```python
import csv

# Define el nombre del archivo CSV original y el nuevo archivo UTF-8
archivo_original = 'datos/datosgob_SIEM2023.csv'
archivo_utf8 = 'datos_utf8.csv'

# Abre el archivo original y el nuevo archivo UTF-8
with open(archivo_original, 'r', encoding='ISO-8859-1') as archivo_original, open(archivo_utf8, 'w', newline='', encoding='utf-8') as archivo_nuevo:
    
    # Lee el contenido del archivo original
    contenido = archivo_original.read()
    
    # Escribe el contenido en el nuevo archivo UTF-8
    archivo_nuevo.write(contenido)

print(f'Archivo convertido a UTF-8 y guardado como {archivo_utf8}')
```
